```python
import pandas as pd
import numpy as np
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from nltk.stem import PorterStemmer
from wordcloud import WordCloud
```

```python
companies = companies[~companies['finalText'].isna()]
```

```python
companies.shape
```

```python
#Tokenize final text, remove stop words and do stemming
companies['finalText'] = companies['finalText'].apply(nltk.word_tokenize)
companies.head()
```

```python
def remove_stop_words(new_text):
    result = [w.lower() for w in new_text if w.lower() not in ENGLISH_STOP_WORDS]
    return result

companies['finalText'] = companies.finalText.apply(remove_stop_words)
companies.head()
```

```python
stemmer = PorterStemmer()

def stem_words(tokens):
    return [stemmer.stem(w) for w in tokens]

companies['finalText'] = companies.finalText.apply(stem_words)
companies.head()
```

```python
def join_tokens(tokens):
    return " ".join(tokens)

companies['finalText'] = companies.finalText.apply(join_tokens)
companies.head()
```

```python
companies.to_csv('CompaniesPostStemming.csv')
```

```python
#Add the data of the same company to create on long string of all filings over the 2 years
tickers = companies['ticker'].unique()
len(tickers)
```

```python
count = 0
companies2 = pd.DataFrame(columns=['ticker', 'text'])
for i in tickers:
    temp = companies[companies['ticker'] == i]
    companies2.loc[count] = [i, temp['finalText'].str.cat(sep = ' ')]
    count+=1
    
```

```python
companies2
```

```python
companies2['count'] = [len(text.split(' ')) for text in companies2['text']]
```

```python
companies2.head()
```

```python
#Now we create the TF-IDF matrix manually 
word_vectorizer = CountVectorizer(token_pattern = '[a-zA-Z]+')

# Now use this to set-up the same type of dataset as before:
vectorizer = word_vectorizer.fit(companies2['text'])
word_bow_matrix = vectorizer.transform(companies2['text'])
count_vectorizer = pd.DataFrame(word_bow_matrix.toarray(), 
                           columns = word_vectorizer.get_feature_names_out())
```

```python
count_vectorizer.head()
```

```python
tf_matrix = count_vectorizer/count_vectorizer.sum(axis = 0)
tf_matrix.head()
```

```python
idf_matrix = tf_matrix.apply(lambda row: (row!=0).sum(), axis = 0)
idf_matrix = np.log(len(tf_matrix)/idf_matrix)
idf_matrix.head()
```

```python
tfidf_matrix = tf_matrix * idf_matrix
tfidf_matrix.head()
```

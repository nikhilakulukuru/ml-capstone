```python
import numpy as np
import pandas as pd
import yfinance as yf
import plotly.express as px
```

## Input
lookback: how many years to lookback  
x: measures in x axis  
y: measures in y axis  
sector: define specific sectors to plot (if input is an empty list, then plot all sectors)

```python
# input
lookback = 3
x = 'Net Income'
y = 'Profit Margin'
sector = ['Industrials']
```

```python
# run function
plot_pp(lookback, x, y, sector)
```

```python
# function definition
def plot_pp(lookback, x, y, sector):
    # read data
    pp_df = pd.read_csv('pp_df.csv')
    # plot
    plot_df = pp_df[pp_df['Year'] > pp_df['Year'].max() - lookback].reset_index(drop=True)
    plot_df['Year'] = plot_df['Year'].astype(str)
    if (len(sector) == 0):
        for ind in plot_df['GICS Sector'].unique():
            fig = px.scatter(plot_df[plot_df['GICS Sector'] == ind], x = x, y = y, color = 'Year', hover_name= 'Symbol')
            fig.update_layout(title = 'Profit Pool of ' + ind,
                          xaxis_title = x,
                          yaxis_title = y,
                          title_x = 0.5,
                          plot_bgcolor='lightgray')
            fig.show()
    else:
        for ind in sector:
            fig = px.scatter(plot_df[plot_df['GICS Sector'] == ind], x = x, y = y, color = 'Year', hover_name= 'Symbol')
            fig.update_layout(title = 'Profit Pool of ' + ind,
                          xaxis_title = x,
                          yaxis_title = y,
                          title_x = 0.5,
                          plot_bgcolor='lightgray')
            fig.show()
```
